```python
from corrector import ColorCorrector
import os
import matplotlib.pyplot as plt
import numpy as np
from imageio import imread
```

```python
cc = ColorCorrector('./checkpoints/scmn_20220925_111650_150.pth')
```

```python
img_path = './test_images/COCO_test2014_000000013377.jpg'
image = imread(img_path) / 255
```

```python
cor_image = cc.correct(image)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Input image")

plt.subplot(1,2,2)
plt.imshow(cor_image)
plt.title("Output image")
```
